```python
from youtube_transcript_api import YouTubeTranscriptApi
from youtube_transcript_api._errors import TranscriptsDisabled, NoTranscriptFound
import os

def extract_text_from_youtube_video(video_id, file_name, drive_folder, language_code="ar"):
    try:
        # Récupérer les sous-titres en arabe
        transcript = YouTubeTranscriptApi.get_transcript(video_id, languages=[language_code])

        # Chemin du fichier de sortie
        output_file_path = os.path.join(drive_folder, file_name)

        # Sauvegarder les sous-titres dans un fichier texte
        with open(output_file_path, 'a', encoding='utf-8') as f:
            for entry in transcript:
                f.write(entry['text'] + '\n')

        print(f"Sous-titres en arabe sauvegardés dans '{output_file_path}'")

    except TranscriptsDisabled:
        print(f"Les sous-titres sont désactivés pour la vidéo : https://www.youtube.com/watch?v={video_id}")
    except NoTranscriptFound:
        print(f"Aucun sous-titre en arabe disponible pour la vidéo : https://www.youtube.com/watch?v={video_id}")
    except Exception as e:
        print(f"Erreur inattendue : {e}")
```
