```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup
import time

# Définir le chemin vers chromedriver
service = Service("/content/chromedriver.exe")  # Remplacez par le chemin vers votre fichier chromedriver
driver = webdriver.Chrome(service=service)

# Charger la page YouTube
driver.get('https://www.youtube.com/@souhailhaddade8406/videos')

# Attendre le chargement de la page
time.sleep(5)  # Ajustez ce délai en fonction de votre connexion

# Récupérer le HTML chargé
soup = BeautifulSoup(driver.page_source, 'html.parser')
driver.quit()

# Extraire les vidéos
for video in soup.find_all('a', href=True):
    if '/watch?' in video['href']:
        title = video.get('title', 'No title available')
        print(f"Titre : {title}\nhref : {video['href']}\n")
```
